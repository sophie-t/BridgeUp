# Look at dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("bird_dino_data.csv")
```

```python
#add a new column for the mass of the whole brain by adding the 
#olfactory bulbs 
#cerebrum 
#optic lobes 
#cerebellum
#brain stem 
df["Whole Brain"]= df["Olfactory bulbs (cm3)"]+df["Cerebrum (cm3)"]+df["Optic Lobes (cm3)"]+df["Cerebellum (cm3)"]+df["Brain Stem (cm3)"]
```

```python
df
```

# Avg BB and CB for birds and dinosaurs

```python
#use .loc function to seperate the bird vs dino column
bird = df.loc[df["Bird vs Dino"]== "B"]
```

```python
bird["Bird Brain/Body mass"]= bird["Whole Brain"] / (bird["Body Mass (kg)"]*1000)
bird["Bird Cerebrum/Brain"] = bird["Cerebrum (cm3)"] / bird["Whole Brain"]
```

```python
bird
```

```python
birdAVG_BB = bird["Bird Brain/Body mass"].mean()
birdAVG_CB = bird["Bird Cerebrum/Brain"].mean()

print(birdAVG_BB)
print(birdAVG_CB)
```

```python
dino = df.loc[df["Bird vs Dino"]== "D"]
```

```python
dino["Dino Brain/Body mass"]= dino["Whole Brain"] / (dino["Body Mass (kg)"]*1000)
dino["Dino Cerebrum/Brain"] = dino["Cerebrum (cm3)"] / dino["Whole Brain"]
```

```python
dino
```

```python
dinoAVG_BB = dino["Dino Brain/Body mass"].mean()
dinoAVG_CB = dino["Dino Cerebrum/Brain"].mean()

print(dinoAVG_BB)
print(dinoAVG_CB)
```

```python
#brain to body mass
unknown_bb = brain_volume_cm3 / (unknown_df_bodymass*1000)
print(unknown_bb)
```

```python
#cerebrum to whole brain
unknown_cb = cerebrum_width_mm/whole_brain_mm
print(unknown_cb)
```
